```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load Iris dataset from sklearn
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the species column with correct mapping
iris_df['species'] = iris.target
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Verify the columns in the DataFrame
print(iris_df.columns)

# Display the first few rows of the dataset
print(iris_df.head())

# Summary statistics
print(iris_df.describe())

# Basic information about the dataset
print(iris_df.info())

```

```python
# Pairplot to visualize relationships
sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"])
plt.show()
```

```python
# Boxplot for each feature
for feature in iris.feature_names:
    sns.boxplot(x='species', y=feature, data=iris_df)
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

```python
# Calculate correlation matrix excluding 'species' column
correlation_matrix = iris_df.drop(columns='species').corr()

# Display correlation matrix
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap of correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
iris_df.to_csv('iris_dataset.csv', index=False)
```
